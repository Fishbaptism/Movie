```python
import pandas as pd
import numpy as np
import csv
```

```python
row = pd.read_csv("C:\\Users\\msi-\\Documents\\rowData.txt", header = None)
row
```

```python
col = pd.read_csv("C:\\Users\\msi-\\Documents\\colData.txt", header = None)
col
```

```python
row.iloc[:, 1]
```

```python
n = 400
user = row.shape[0]
item = col.shape[1]
user, item
```

```python
row = np.array(row).astype(bool)
col = np.array(col).astype(bool)
row.shape, col.shape
```

```python
rowZip = []
tmp = np.arange(user)
for i in range(n):
    rowZip.append(tmp[row[:, i]])
rowZip
```

```python
colZip = []
tmp = np.arange(item)
for i in range(n):
    colZip.append(tmp[col[i, :]])
colZip
```

```python
max([max(i) for i in colZip])
```

```python
','.join(colZip[1].astype(str))
```

```python
with open('rowZip.csv', 'w', newline='') as file:
    for i in range(n):
        file.write(','.join(rowZip[i].astype(str)) + '\n')
```

```python
with open('colZip.csv', 'w', newline='') as file:
    for i in range(n):
        file.write(','.join(colZip[i].astype(str)) + '\n')
```
